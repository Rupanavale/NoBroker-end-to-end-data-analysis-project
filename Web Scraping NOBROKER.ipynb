```python
soup = BeautifulSoup(html_text.content,'lxml')
```

```python
data = []

# Extract all property listings
listings = soup.find_all('article', {'aria-label': 'article'})

for listing in listings:
    # Extract property details
    property_title = listing.find('h2', class_ = 'heading-6').find('a').text.strip()
    property_location = listing.find('div',class_ = 'mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95').text
    property_rent = listing.find('div', class_ = 'flex').find_next('div',id="minimumRent").text
    property_deposit = listing.find('div',class_ = 'flex' ).find_next('div',id = "roomType").text
    property_area = listing.find('div', itemprop= 'additionalProperty').find('div', class_='font-semi-bold').text.strip()
    property_type = listing.find('div', class_ = 'flex-1 border-r border-r-solid border-r-cardbordercolor flex').find('div', class_ ='font-semibold').text
    property_parking = listing.find('div', class_ = 'flex flex-1 pl-0.5p').text
    property_furnishing = listing.find('div',class_ = 'flex flex-1 border-r border-r-solid border-r-cardbordercolor').text
    property_available_from = listing.find('div', class_ ='flex p-0.5p items-center').find_next('div',class_ = 'flex flex-1 pl-0.5p').text
    
    
    print(f"Property Title: {property_title}")
    print(f"Location: {property_location}")
    print(f"Rent: {property_rent}")
    print(f"Deposit: {property_deposit}")
    print(f"Area: {property_area}")
    print(f"Type: {property_type}")
    print(f"Parking: {property_parking}")
    print(f"Furnishing: {property_furnishing}")
    print(f"Available From: {property_available_from}")
    
    print(" ")
    
    data.append([property_title, property_location, property_rent, property_deposit, property_area, property_type, property_parking, property_furnishing,property_available_from])
    
    df = pd.DataFrame(data, columns=['Property Title','Location','Rent','Deposit','Area','Type','Parking','Furnishing','Available From'])             

    df.to_excel('nobroker.xlsx')
    
    
```
